```python
# import necessary libraries

import matplotlib.pyplot as plt
from sklearn import datasets
import xgboost as xgb
```

```python
# load and visualize data

digits = datasets.load_digits()

X = digits.data
y = digits.target
images=digits.images
plt.imshow(images[0],cmap='binary',interpolation='nearest')
```

```python
targets=digits.target
print(targets[0])
```

```python
images=images.reshape(1797,8*8)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(images, targets, test_size=0.2)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

#accuracy = accuracy_score(y_test, y_pred)
#print("Accuracy: %.2f%%" % (accuracy * 100.0))
print("accuracy score:{:.4f}".format(accuracy_score(y_test, y_pred)))
print("Classification Results:\n{}".format(classification_report(y_test, y_pred)))
confusion = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=confusion)
disp.plot(cmap=plt.cm.Blues, xticks_rotation=45)
```

## KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier

KNN = KNeighborsClassifier(n_neighbors=1)
KNN.fit(X_train, y_train)
y_pred = KNN.predict(X_test)
print("accuracy score:{:.4f}".format(accuracy_score(y_test, y_pred)))
print("Classification Results:\n{}".format(classification_report(y_test, y_pred)))
confusion = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=confusion)
disp.plot(cmap=plt.cm.Blues, xticks_rotation=45)
```

## Decision Tree

```python
DTC = DecisionTreeClassifier()
DTC.fit(X_train, y_train)
y_pred = DTC.predict(X_test)
print("accuracy score:{:.4f}".format(accuracy_score(y_test, y_pred)))
print("Classification Results:\n{}".format(classification_report(y_test, y_pred)))
```

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=confusion)
disp.plot(cmap=plt.cm.Blues, xticks_rotation=45)
print('Confusion Matrix\n')
print(confusion)
```
